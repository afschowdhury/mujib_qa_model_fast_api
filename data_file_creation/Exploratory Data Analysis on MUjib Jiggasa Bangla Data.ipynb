## Defining Plotting Function

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.style.available

plt.style.use(plt.style.available[4])
```

```python
def plot_bar_plot(x,y,xlabel,ylabel,title,show_number=False,show_x=False,rotation='vertical',t_f_s = 20,figsize=(26,8)):
    
    
    
    fig , ax = plt.subplots(figsize=figsize)
    
    x = x
    y = y
    
    colors = plt.cm.Dark2(range(6))
    
    # prop = fm.FontProperties(fname='Nikosh.ttf')
    prop = fm.FontProperties(fname='data/fonts/Nikosh.ttf')


    bar_plot = ax.bar(x,y,color = colors)

    ax.set(xlabel= xlabel,
        ylabel = ylabel);
    
    if show_x:
        ax.set_xticks(ticks = x, fontproperties=prop,labels= x,rotation=rotation)#,fontproperties=font) 
    
    if show_number:
        ax.bar_label(bar_plot,padding=3)
    
    fig_name = "plots/"+title+".png"
        
    # fig.savefig(fig_name)
    fig.suptitle(t=title,
             size = t_f_s)
    fig.savefig(fig_name)
```

## Questions per Tags

```python
t1 = list(df.T1.unique())
t1
```

```python
t2 = list(df.T2.unique())
t2
```

```python
tags = []

tags.extend(t1)
tags.extend(t2)

len(tags), len(t2)
```

```python
c_tags = []

for i in tags:
    if i not in c_tags:
        c_tags.append(i)

len(c_tags), len(tags)
```

```python
tags_q_dict = dict()

for i in c_tags:
    if type(i) != str:
        text = "tags_q_dict['missing'] = []"
        exec(text)
    else:
        text = "tags_q_dict['"+i+"'] = []"
        exec(text)
```

```python
tags_q_dict
```

```python
for tag in tags_q_dict.keys():
    for item in data:
        for  qas in  item['qas']:
            for t in qas['tags']:
                if type(t) != str:
                    t = "missing"
                    # continue
                if t == tag:
                    tags_q_dict[tag].append(qas['qtns'])
                    
```

```python
q_per_tags = dict()

for tag in tags_q_dict.keys():
    if tag == 'missing':
        continue
    q_per_tags[tag] = len( tags_q_dict[tag])
q_per_tags
```

```python
a = list(q_per_tags.keys())
b = list(q_per_tags.values())
plot_bar_plot(a,
              b,
              "Tags",
              "Questions",
              "Questions per Tag",
              show_number=True,
              show_x=True)
```
